```python
import pandas as pd
from sklearn.datasets import load_digits
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
import json
```

```python
with open('model-config.json','r') as f:
    modelsConfig=json.load(f)
```

```python
digits=load_digits()
```

```python
import sklearn
```

```python
scores=[]
for modelName,modelConfig in modelsConfig.items():
    clf=GridSearchCV(eval(modelConfig["model"]),modelConfig["param"],cv=5,return_train_score=False)
    clf.fit(digits.data,digits.target)
    scores.append({
        'model':modelName,
        'score':clf.best_score_,
        'param':clf.best_params_
    })
```

```python
scores
```

```python
modelScoreDf=pd.DataFrame(scores,columns=['model','score','param'])
```
