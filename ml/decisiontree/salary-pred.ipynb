```python
import pandas as pd
```

```python
df=pd.read_csv('salary.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
df['le_company']=le_company.fit_transform(df.company)
df['le_job']=le_job.fit_transform(df.job)
df['le_degree']=le_degree.fit_transform(df.degree)
```

```python
dfEnc=df.drop(['company','job','degree'],axis='columns')
```

```python
dfEnc
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(dfEnc.drop(columns='salary_more_then_100k'),dfEnc.salary_more_then_100k,train_size=.75)
```

```python
from sklearn import linear_model
from sklearn import tree
```

```python
model=linear_model.LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.score(X_train,y_train)
```
